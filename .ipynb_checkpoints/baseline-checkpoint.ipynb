#  generate train dataframe

```python
import numpy as np
import pandas as pd
from scipy.sparse import csr_matrix, load_npz, hstack
```

```python
import import_ipynb
from loadData import raw_df
```

```python
import lightgbm as lgb
```

```python
item_property_matrix = load_npz('../processed_data/item_property.npz')
need_one_hot_columns_matrix = load_npz('../processed_data/need_onehot_columns.npz')
item_third_level_matrix = pd.read_csv('../processed_data/item_third_level_one_hot.csv', sep=' ').as_matrix()
time_matrix = pd.read_csv('../processed_data/time.csv', sep=' ').as_matrix()
```

#  without scaled coulmns

```python
_columns = ['item_price_level','item_sales_level','item_collected_level','item_pv_level',
                     'user_age_level','user_occupation_id','user_star_level','context_page_id',
                     'shop_review_num_level','shop_review_positive_rate','shop_star_level','shop_score_service',
                    'shop_score_description', 'day', 'hour']
without_scaled_matrix = raw_df[_columns].as_matrix()
```

```python
predict_category_matrix = load_npz('../processed_data/predict_categories.npz')
```

```python
ndarray_matrix = np.concatenate((without_scaled_matrix,np.concatenate((item_third_level_matrix, time_matrix), axis=1)), axis=1)
```

```python
sparse_matrix = hstack((hstack((item_property_matrix, need_one_hot_columns_matrix)), predict_category_matrix))
totalData = hstack((csr_matrix(ndarray_matrix), sparse_matrix))
```

```python
totalData.shape
```

## generate train and test data

```python
train_Y, test_Y = raw_df[raw_df.day != 24]['is_trade'], raw_df[raw_df.day == 24]['is_trade']
```

```python
train_X, test_X = totalData.tocsr()[:420693, : ], totalData.tocsr()[420693:]
```

```python
# valid_Y = raw
```

```python
gbm = lgb.LGBMClassifier(objective='binary',
                 num_leaves=2048,
                 learning_rate=0.01,
                 n_estimators=2000,
                 max_depth=7,
                 min_samples_split=100)
evals_result = {}
gbm.fit(train_X, train_Y.values,
#         feature_name = list(trainX.columns.values),
        eval_set=[(test_X, test_Y.values)],
        eval_metric='binary_logloss',
        early_stopping_rounds=100
       )
```
